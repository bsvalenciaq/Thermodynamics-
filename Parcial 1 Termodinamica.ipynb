## 1. Un fluido está contenido en un cilindro por un pistón sin fricción cargado por resorte, de modo que la presión en el fluido es una función lineal del volumen $p = a + bV$. La energía interna del fluido viene dada por la ecuación, $U = 45 + 3.5pV$, donde U está en kJ, p en kPa y V en metro cúbico. Si el fluido cambia de un estado inicial de $200 kPa$, $0.05 m^3$ a un estado final de $420 kPa$, $0.09 m^3$, sin más trabajo que el realizado en el pistón, encuentre la dirección y la magnitud del trabajo y la transferencia de calor.

**Solucion**

$U=45+3.5pV$  

$U_1=45+3.5p_1V_1$

$U_2=45+3.5p_2V_2$

$p_1=2x10^5$, $p_2=4.2x10^5$,  $V_1=0.05m^3$, $V_2=0.09m^3$ 



$ΔU=(U_2-U_1)$ = $(45+3.5p_2V_2-45+3.5p_1V_1)$ = $3.5*(p_2V_2-p_1V_1)$

$ΔU=(4.2x10^5*0.09-2x10^5*0.05)$ = $97300J=97.3kJ$

p se da como:

$p = a + bV$; $p_1 = a + bV_1$; $p_2 = a + bV_2$

$200=a+b*0.05$ (i)

$420=a+b*0.09$ (ii)

Restando (i) a (ii)

$220=0.04*b$

$b=\frac{220}{0.04}$ = $5500kN/m^2$

Reemplazando b en (i) obtenemos 

$200=a+5500*0.05$

$a=200-5500*0.05$ = $-75$

Transferencia de trabajo durante el proceso 

$W_{1-2}=\int_{v_1}^{v_2} pdV$ = $\int_{v1}^{v2} (a+bV)dV$ = $a(v_2-v_1)+b(\frac{v_2^2-v_1^2}{2})$

$-75*(0.09-0.05)+5500*(\frac{0.09^2-0.05^2}{2})=12.4kJ$







Transferencia de calor durante el proceso

$Q_{1-2}=ΔU+W_{1-2}=(U_2-U_1)+W_{1-2}$ = $ 97.3+12.4=109.7kJ$

El signo positivo es incompatible con el problema actual, por lo que no se considera
